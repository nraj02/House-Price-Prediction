```python
x = df.filter(['sqft_living','view','sqft_basement','bedrooms','sqft_above','bathrooms','floors'])
y = df.filter(['price'])
```

```python
from sklearn.model_selection  import train_test_split
x_train,x_test,y_train,y_test = train_test_split(x ,y ,test_size=0.2, random_state=42)
```

```python
from sklearn.preprocessing import StandardScaler
sc_X = StandardScaler()
x_train = sc_X.fit_transform(x_train)
x_test = sc_X.transform(x_test)
y_train = sc_X.fit_transform(y_train)
y_test = sc_X.transform(y_test)

sc_y = StandardScaler()
y_train = sc_y.fit_transform(y_train.reshape(-1,1)).flatten()
```

```python
from xgboost import XGBRegressor
lr = XGBRegressor()
lr.fit(x_train,y_train)
```

```python
from sklearn.metrics import r2_score
from sklearn.metrics import mean_squared_error
y_pred = lr.predict(x_test)
r2_lr_test = r2_score(y_test, y_pred)
print("R2 Test Score:", r2_lr_test)
mse_lr_test = mean_squared_error(y_test, y_pred)
print("Mean Squared Error of Test:", mse_lr_test)
```
